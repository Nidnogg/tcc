```python
import pandas as pd
```

```python
df = pd.read_csv("runtimes_min_max.csv")
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(10,6)})
ax = sns.boxplot(data=df[df.model != 'test_ada'], x='runtimes', y='model')
ax.set(title='Overview of Min Max runtimes', ylabel='Classifiers', xlabel='Runtimes', xlim=(0, 37))

ax

plt.savefig('./plots/min_max_all.png', bbox_inches='tight')

```

```python
sns.set(rc={'figure.figsize':(10,0.5)})
ax = sns.boxplot(data=df[df.model == 'test_ada'], x='runtimes', y='model')
ax.set(title='Overview of Min Max runtimes', ylabel='Classifiers', xlabel='Runtimes')

ax

plt.savefig('./plots/min_max_ada.png', bbox_inches='tight')
```
